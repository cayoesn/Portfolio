# K-means (Moons)


## Importação das bibliotecas


```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

## Importação dos valores do eixo vertical e horizontal da base de dados make_moons


```python
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
```

## Visualização dos dados


```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

## Clusterização com o algoritmo K-means


### Com k = 2

```python
kmeans = KMeans(n_clusters=2)
kmeans = kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

### Com k = 3

```python
kmeans = KMeans(n_clusters=3)
kmeans = kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

## Clusterização com o algoritmo DBSCAN


```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], c="yellow",
            marker="o", s=40, label="cluster 1")

plt.scatter(X[y == 1, 0], X[y == 1, 1], c="green",
            marker="o", s=40, label="cluster 2")

plt.legend(('Cluster 1', 'Cluster 2'), loc=1)

plt.show()
```
