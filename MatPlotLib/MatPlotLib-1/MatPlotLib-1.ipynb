# MatPlotLib


```python
import matplotlib.pyplot as plt

%matplotlib inline
```

## Definindo o diagrama

```python
values = [1, 5, 8, 9, 2, 0, 3, 10, 4, 7]
plt.plot(range(1, 11), values)
plt.show()
```

## Desenhando várias linhas e diagramas

```python
values = [1, 5, 8, 9, 2, 0, 3, 10, 4, 7]
values2 = [3, 8, 9, 2, 1, 2, 4, 7, 6, 6]
plt.plot(range(1, 11), values)
plt.plot(range(1, 11), values2)
plt.show()
```

## Formatando os eixos

```python
values = [0, 5, 8, 9, 2, 0, 3, 10, 4, 7]
ax = plt.axes()
ax.set_xlim([0, 11])
ax.set_ylim([-1, 11])
ax.set_xticks(range(1, 11))
ax.set_yticks(range(0, 11))
plt.plot(range(1, 11), values)
plt.show()
```

## Acrescentando as grades

```python
values = [0, 5, 8, 9, 2, 0, 3, 10, 4, 7]
ax = plt.axes()
ax.set_xlim([0, 11])
ax.set_ylim([-1, 11])
ax.set_xticks(range(1, 11))
ax.set_yticks(range(0, 11))
ax.grid()
plt.plot(range(1, 11), values)
plt.show()
```

## Rótulos, anotações e legendas

```python
values = [1, 5, 8, 9, 2, 0, 3, 10, 4, 7]
values2 = [3, 8, 9, 2, 1, 2, 4, 7, 6, 6]
plt.xlabel('Entries')
plt.ylabel('Values')
plt.annotate(text='First Entry', xy=[1, 1])
line1 = plt.plot(range(1, 11), values)
line2 = plt.plot(range(1, 11), values2)
plt.legend(('First', 'Second'), loc=4)

plt.show()
```
